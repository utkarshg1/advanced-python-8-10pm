```python
from dataclasses import dataclass
from typing import Literal
```

```python
@dataclass
class Person2:
    
    name: str
    age: int
    gender: Literal["male", "female"]
    occupation: str

    def __post_init__(self):
        errors = []

        if not isinstance(self.name, str):
            errors.append("name should be a string")
        if not isinstance(self.age, int) or self.age <= 0:
            errors.append("age should be a positive integer")
        if self.gender not in ["male", "female"]:
            errors.append("gender can be either male or female")
        if not isinstance(self.occupation, str):
            errors.append("occupation should be a string")
        
        if errors:
            raise ValueError("\n".join(errors))
        
    def introduce(self):
        print(f"Hi There my name is {self.name}")
        print(f"I am {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as a {self.occupation}")
```

```python
p4 = Person2(name= 23, age="Test", gender=False, occupation=1.4)
```

```python
p4 = Person2(name="Raman", age=-34, gender="male", occupation="Sales")
```

```python
p4 = Person2(name="Aditi", age=25, gender="female", occupation="Purchase")
```

### Create a bank account class

```python
@dataclass
class Account:

    ac_no: int
    name: str 
    bal: float

    def deposit(self, amt):
        self.bal = self.bal + amt
        print(f"Amount of {amt} INR Deposited in ac no - {self.ac_no}")
        print(f"Current Balance is : {self.bal} INR")

    def withdraw(self, amt):
        if amt > self.bal:
            raise ValueError(f"Insuffecient Balance, Current balance is {self.bal} INR")
        else:
            self.bal = self.bal - amt 
            print(f"Amount of {amt} INR Withdrawn from ac no - {self.ac_no}")
            print(f"Current Balance is : {self.bal} INR")

    def transfer(self, amt, other_ac):
            self.withdraw(amt)
            other_ac.deposit(amt)
```
