### Creating class triangle

```python
from dataclasses import dataclass
from typing import Union
```

```python
@dataclass
class Triangle:
    a: Union[int, float]
    b: Union[int, float]
    c: Union[int, float]

    def __post_init__(self):
        errors = []

        if not isinstance(self.a, (int, float)) or self.a <= 0:
            errors.append("Side a should be a positive integer or float")
        if not isinstance(self.b, (int, float)) or self.b <= 0:
            errors.append("Side b should be a positive integer or float")
        if not isinstance(self.c, (int, float)) or self.c <= 0:
            errors.append("Side c should be a positive integer or float")
        if not self.is_triangle():
            errors.append("Given sides do not form a triangle")

        if errors:
            raise ValueError("\n".join(errors))

    def is_triangle(self):
        if (
            (self.a + self.b > self.c)
            and (self.b + self.c > self.a)
            and (self.c + self.a > self.b)
        ):
            return True
        else:
            return False

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        a = (s * (s - self.a) * (s - self.b) * (s - self.c)) ** (1 / 2)
        return a
```

### Practical application - Calculating area or perimeter of multiple shapes

```python
@dataclass
class Rectangle:
    width: Union[int, float]
    height: Union[int, float]

    def perimeter(self):
        return 2 * (self.width + self.height)

    def area(self):
        return self.width * self.height
    
```

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius: Union[int, float]

    def perimeter(self):
        return 2 * pi * self.radius
    
    def area(self):
        return pi * (self.radius ** 2)
    
```

```python
r1 = Rectangle(20, 10)
r1
```

```python
type(r1)
```

```python
r1.perimeter()
```

```python
r1.area()
```
